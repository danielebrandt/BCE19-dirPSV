# Directional results from data:

Following Brandt et al. (2019) the directional results from data or simulated data are


$<\sigma_x> = \sqrt{ \frac{1}{N-1}\sum{(x - <x>)^2}} , <x> = \frac{1}{N}\sum{x}$


$<\sigma_y> = \sqrt{\frac{1}{N-1}\sum{(y - <y>)^2}} , <y> = \frac{1}{N}\sum{y}$


$ <E'> = <\sigma_y>/<\sigma_x> $
 

$ <A_{dir}> = \sqrt{{(<\sigma_y>.<\sigma_x>)}^2 - {cov(x,y)}^2}$ 

The functions used for calculate them are:

**simGGP.Med_desvxy_sim**

input listaxy (one array with two columns x,y equal area coordinated of rotated directions)

output $<x>$ , $<y>$, $<\sigma_x>$,$<\sigma_y>$


**simGGP.Adir**

input listaxy (one array with two columns x,y equal area coordinated of rotated directions)

output $A_{dir}$

**simGGP.Elon_linha_sim(lista_xy)**

input listaxy (one array with two columns x,y equal area coordinated of rotated directions)

output $E'$

For find confidence regions using NB bootstraping:

**simGGP.mean_boot(x,NB)**


**simGGP.std_boot(x, NB)**



**simGGP.Adir_boot(lista_xy, NB)**



**simGGP.Elon_linha_boot(lista_xy, NB)**




### The x and y coordinates are $x_E$ and $x_N$ from Brandt et al. (2019)

```python


meanx, meany, stdx, stdy = simGGP.Med_desvxy_sim(lista_xy)

Eprime = simGGP.Elon_linha_sim(lista_xy)

Adir = simGGP.Adir(lista_xy)


```

```python
print('Results:')

print('Mean x = %.3f'%(meanx))

print('Std x = %.3f'%(stdx))

print('Mean y = %.3f'%(meany))

print('Std y = %.3f'%(stdy))

print('E = %.2f'%(Eprime))

print('Adir = %.5f'%(Adir))
```

```python
# number of bootstraped re-sampling
# NB

NB = 50
xmin,xmax = simGGP.mean_boot(lista_xy[:,0],NB)
ymin,ymax = simGGP.mean_boot(lista_xy[:,1],NB)

stdxmin,stdxmax = simGGP.std_boot(lista_xy[:,0],NB)
stdymin,stdymax = simGGP.std_boot(lista_xy[:,1],NB)

Emin,Emax = simGGP.Elon_linha_boot(lista_xy, NB)

Amin,Amax = simGGP.Adir_boot(lista_xy, NB)
```

```python
print('Results:')

print('Mean x = %.3f, limits of confidence [min,max]: [%.3f,%.3f]'%(meanx,xmin,xmax))

print('Std x = %.3f, limits of confidence [min,max]: [%.3f,%.3f]'%(stdx,stdxmin,stdxmax))

print('Mean y = %.3f, limits of confidence [min,max]: [%.3f,%.3f]'%(meany,ymin,ymax))

print('Std y = %.3f, limits of confidence [min,max]: [%.3f,%.3f]'%(stdy,stdymin,stdymax))

print('E = %.2f, limits of confidence [min,max]: [%.2f,%.2f]'%(Eprime,Emin,Emax))

print('Adir = %.5f, limits of confidence [min,max]: [%.5f,%.5f]'%(Adir,Amin,Amax))
```
